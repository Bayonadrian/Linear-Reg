### Dealing with outliers
One way to deal with **outliers** is to remove the largest 1% of data.

```python
# The 99th quantile is sought, the one with 99% of the data below

q = df['Price'].quantile(0.99)
q
```

```python
# Now the maximum data is 129222.000000 so that the curve will not get lost in skews

mod_df = df[df['Price'] < q]
mod_df.describe(include='all')
```

```python
# Note that the graph looks much clearer without so much noise

sns.distplot(mod_df['Price'])
plt.show()
```

#### Mileage

```python
# The 99th quantile is sought, the one with 99% of the data below

q = mod_df['Mileage'].quantile(0.99)
q
```

```python
mod_df = mod_df[mod_df['Mileage'] < q]
mod_df.describe(include='all')
```

```python
sns.distplot(mod_df['Mileage'])
plt.show()
```

#### EngineV

```python
# In this case, 6.5 is used because it is the standard in motors

mod_df = mod_df[mod_df['EngineV'] < 6.5]
mod_df.describe(include='all')
```

```python
sns.distplot(mod_df['EngineV'])
plt.show()
```

#### Year

## Assumptions
It is necessary to assume certain criteria before starting to do a linear regression.

### Linearity
The dependent variable is **Price** which corresponds to the price of the vehicle, so it is necessary to see if it is linear with the rest of the variables.

```python
# Note that no distribution has normal form so one or more variables must be modified.

f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3))
ax1.scatter(mod_df['Year'],mod_df['Price'])
ax1.set_title('Price and Year')
ax2.scatter(mod_df['EngineV'],mod_df['Price'])
ax2.set_title('Price and EngineV')
ax3.scatter(mod_df['Mileage'],mod_df['Price'])
ax3.set_title('Price and Mileage')
```

```python
# The Price variable does not have a normal distribution unlike the rest.
# So the relationship cannot be linear

fig, axs = plt.subplots(ncols=4, figsize=(15,3))
sns.distplot(mod_df['Price'], ax=axs[0])
sns.distplot(mod_df['Year'],ax=axs[1])
sns.distplot(mod_df['Mileage'],ax=axs[2])
sns.distplot(mod_df['EngineV'],ax=axs[3])
plt.show()
```

#### Logarithmic transformation
The logarithmic transformation is very useful to be able to transform a distribution with certain characteristics into a normal distribution. This can be applied using the most common logarithms such as **natural logarithm**, **base 2 logarithm** and **base 10 logarithm**, using these it is possible to make said transformation. Intuitively **by applying a base 10 logarithm it is possible to reduce the zeros to numbers by 10, so that if the figure has 4 digits it becomes a number close to 4**. This is how it works with all other logarithms.

```python
import numpy as np

np.log10(15627)
```

```python
# One way to solve this is to obtain the logarithm of Price.

log_price = np.log(mod_df['Price'])
mod_df['log_price'] = log_price
mod_df
```

```python
# The distribution is normalized.

sns.distplot(mod_df['log_price'])
```

```python
# The distributions are made linear

f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3))
ax1.scatter(mod_df['Year'],mod_df['log_price'])
ax1.set_title('Price and Year')
ax2.scatter(mod_df['EngineV'],mod_df['log_price'])
ax2.set_title('Price and EngineV')
ax3.scatter(mod_df['Mileage'],mod_df['log_price'])
ax3.set_title('Price and Mileage')
```

### No endogeneity
It refers to the correlation between the dependent variable (**y**) and the errors, this is due to the lack of one or more important variables for the model, as an example we can consider a study in London departments which for some reason have an inverse price-size relationship, which is counterintuitive, this happened because the apartments closest to the center are the most expensive and smallest, so the model does not have a very important variable, which is the location, and this fact can produce an inefficient and erroneous model.

### Normality homoelasticity
- Normality, it is possible to define that the data has a normal distribution.
- Zero mean, it is possible to define that this concept is applicable because the intercept exists
- Homoelasticity, the data is distributed homogeneously within the plots, otherwise the data was transformed into a logarithmic expression, so that the inference is correct.

---

This concept usually occurs when the data is compact on the one hand and over time begins to disperse, which makes future predictions incorrect.

### No autocorrelation
The data does not come from a series of some kind, so there is no autocorrelation.

### Multicollinearity
It is possible that there are two closely related variables in the model, the **Year** variable and the **Mileage** variable. It is normal for the mileage to be higher in relation to the years.

```python
# It is possible to see multicollinearity through a correlation.

mod_df.corr()
```

```python
# A VIF greater than 10 is too high, for this reason it is understood that the Year variable should be removed.

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = {'Mileage': 0, 'Year': 0, 'EngineV': 0}

vif['Mileage'] = variance_inflation_factor(mod_df[['Mileage','Year','EngineV']].values, 0)
vif['Year'] = variance_inflation_factor(mod_df[['Mileage','Year','EngineV']].values, 1)
vif['EngineV'] = variance_inflation_factor(mod_df[['Mileage','Year','EngineV']].values, 2)

pd.DataFrame(vif, index=['Data']).transpose()
```

```python
mod_df.drop(['Price', 'Year'], axis=1, inplace=True)
mod_df.head()
```

## Dummies

To obtain dummy variables, it is necessary to remove the element from the beginning, because maintaining this element generates **multicollinearity** since being all 0 necessarily involves the first element implicitly. But before removing this element it is necessary to know which element to remove, for more information see the following link https://towardsdatascience.com/drop-first-can-hurt-your-ols-regression-models-interpretability-4ca529cfb707.
In turn, once the first element has been removed, which in this case is **Audi**, in case you want to use this element, it is necessary to put 0 in all the others, this is equivalent to **Audi**.

```python
# As a test it is recommended not to use the drop_frist parameter and use elk VIF analysis
# To do the comparison.

dummiesVar = pd.get_dummies(mod_df, drop_first=True)
dummiesVar.head()
```

### Reordering the dataset

```python
dummiesVar.columns
```

```python
cols = ['log_price', 'Mileage', 'EngineV', 'Brand_BMW', 'Brand_Mercedes-Benz',
       'Brand_Mitsubishi', 'Brand_Renault', 'Brand_Toyota', 'Brand_Volkswagen',
       'Body_hatch', 'Body_other', 'Body_sedan', 'Body_vagon', 'Body_van',
       'Engine Type_Gas', 'Engine Type_Other', 'Engine Type_Petrol',
       'Registration_yes']
```

```python
preprocessed_data = dummiesVar[cols]
preprocessed_data.head()
```

# Linear regression

## Defining variables
The variables **X** and **y** are defined.

```python
y = preprocessed_data['log_price']
X_data = preprocessed_data.drop('log_price', axis=1)
```

## Scaling variables
All variables within **X_data** need to be scaled.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X_data)
X
```

## Data selection
You need to split the data between **train** and **test**.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=365)
```

## Creating the regression

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

### Analyzing the regression

#### Relational analysis

```python
# A prediction is made using the training data.

y_hat = reg.predict(x_train)

# Plot the prediction against the actual data to see if they are close
# Note that they are not exact but they are similar and have linearity

plt.scatter(y_train, y_hat)
plt.xlabel('Targets (y_train)',size=18)
plt.ylabel('Predictions (y_hat)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

#### Residual analysis


```python
# Of the assumptions, the errors must have a normal distribution and a mean of 0
# The tail on the right side means that the model overestimates low values

sns.distplot(y_train - y_hat)
plt.title("Residuals PDF", size=18)
```

#### R cuadrado

```python
# The model has the capacity to infer over 74% of the variability of the data.

reg.score(x_train,y_train)
```

#### Intercept

```python
reg.intercept_
```

#### Coefficients

```python
reg.coef_
```

```python
# The slopes, both positive and negative, determine the result of the curve
# that in case of numeric variables
# But with categorical variables things change a bit, you have to remember that an element was deleted
# which was Audi, this becomes the reference point, so that when BMW is greater than 0
# it can be inferred that it is more expensive than the Audi and since Toyota is less than 0 then it is cheaper
# That the Audi.

reg_summary = pd.DataFrame(X_data.columns.values, columns=['Variables'])
reg_summary['Pendientes'] = reg.coef_
reg_summary
```

### Using tests

```python
y_hat_test = reg.predict(x_test)
```

```python
# Note that the model is better at predicting models with high prices than those with low prices

plt.scatter(y_test, y_hat_test)
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
# To have a better vision of where more data is concentrated, the alpha parameter is used
# And it is observed that the highest concentration is in the middle

plt.scatter(y_test, y_hat_test, alpha=0.2)
plt.xlabel('Targets (y_test)',size=18)
plt.ylabel('Predictions (y_hat_test)',size=18)
plt.xlim(6,13)
plt.ylim(6,13)
plt.show()
```

```python
# For a better review of the predictions it is necessary to bring the data of their expressions
# logarithmic to their real expressions, this is achieved with the exponential or np.exp

df_pf = pd.DataFrame(np.exp(y_hat_test), columns=['Prediccion'])
df_pf.head()
```

Once the first data is normalized, it is possible to make a better comparison between the predicted data and the real data of the test.

```python
df_pf['Reales'] = np.exp(y_test)
df_pf
```

It is possible to verify that there is a problem so it is better to review a separate post.

```python
# Note that y_test kept the indices it had so that it hinders the visualization.

y_test
```

```python
# The need to reset the index appears

y_test = y_test.reset_index(drop=True)
y_test.head()
```

```python
df_pf['Reales'] = np.exp(y_test)
df_pf
```

```python
# To make a better review it is possible to create two columns in which the residuals can be seen
# and even their percentage differences

df_pf['Residuales'] = df_pf['Reales'] - df_pf['Prediccion']
df_pf['Differencia %'] = np.absolute(df_pf['Residuales']/df_pf['Reales']*100)
df_pf
```

```python
# Note that within the quartiles the differences are not as large.

df_pf.describe()
```

```python
# Now making some adjustments it is possible to see the data manually.
# Note that the data at the end are the ones with the largest differences
# It may be possible to improve the model by adding more data.

pd.options.display.max_rows = 999
pd.set_option('display.float_format', lambda x: '%.2f' % x)
df_pf.sort_values(by=['Differencia %'])
```
